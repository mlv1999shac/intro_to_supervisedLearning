```python
import random
import warnings
warnings.filterwarnings("ignore")

random.seed(100)
```

```python
#lets two features of our data set be 
#grade and 
#bummpy
size=1000
 
roughness =  [ random.random() for i in range(0,size) ]
hillslope  = [ random.random() for i in range(0,size) ]
error  = [ random.random() for i in range(0,size) ]

#y could be either 0 or 1

#if we round then any value less than <= .5 , y=0
#value >.5 gives 1
y = [ round(hillslope[i]*roughness[i]+0.3+0.1*error[i])  for i in range(0,size)]


for i in range(0,len(y)):
    if hillslope[i] > 0.8 or roughness[i] > 0.8:
        y[i] = 1 
#split the data into training set and test set

#zip is a container and that holds the file

X = [[a,b] for a,b in zip(hillslope,roughness)]
#this is lists of lists , each list has two elements one belongs to bumpy and next is grade
#so it a 10*2 form of a matrix like with 10 rows , 10 datapoints and 2 features

#output is y which is only one 

split = int(0.75*size)
X_train = X[0:split]
X_test  = X[split:]

y_train = y[0:split]
y_test  = y[split:]



#now figure out which points where car can go fast and slow
#obviusly those datapoints where car goes slow is where grade is greater and bumpy is greater
#y=1 imples slow , y=0 imples fast

roughness_slow = [ X_train[i][0] for i in range(0,len(X_train)) if y[i]==1 ]
roughness_fast = [ X_train[i][0] for i in range(0,len(X_train)) if y[i]==0 ]

hillslope_slow = [ X_train[i][1] for i in range(0,len(X_train)) if y[i]==1 ]
hillslope_fast = [ X_train[i][1] for i in range(0,len(X_train)) if y[i]==0 ]
    
    

        
roughness_slow
```
